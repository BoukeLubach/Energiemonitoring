```python
# -*- coding: utf-8 -*-
"""
Created on Sun Feb  9 17:39:42 2020

@author: WaterandEnergy
"""
import chart_studio.plotly as py
import pandas as pd
import numpy as np
import mysql.connector
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import time


start_time = time.time()

def get_sql_data(taglist, tablename):   
    """Takes a list of strings with tags and a tablename, 
    returns a pandas dataframe with the tag timeseries"""

    tagstring = taglist[0]
    for nr in range(1, len(taglist)):
        tagstring = tagstring + ', ' + taglist[nr]
        
    print(tagstring)
    conn = mysql.connector.connect(user='root', 
                                  password='pugliawijnislekker',
                                  host='127.0.0.1',
                                  database='Teijin_2019')
    
      
    sqlcommand = 'SELECT Datetime, ' + tagstring + ' FROM ' + tablename
    df = pd.read_sql(sqlcommand, con=conn)
    return df



print("--- %s seconds ---" % (time.time() - start_time))
```

```python
## constants

steamcost = {'MP' : 20,
            'LP' : 17,
            'condensate': 2.5}

MP_steam_29barG =    {'energy':2542,  
                      'cost': steamcost['MP'] - steamcost['condensate']} 
MP_steam_11barG =    {'energy':2412,  
                      'cost': steamcost['MP'] - steamcost['condensate']} 
MP_steam_6barG =     {'energy':2386,  
                      'cost': steamcost['MP'] - steamcost['condensate']} 
LP_steam_3barG =     {'energy':2362,  
                      'cost': steamcost['LP'] - steamcost['condensate']} 
LP_steam_injection = {'energy':2739,  
                      'cost': steamcost['LP']} 


```

```python
##Budget 2400
cols= ['A24F06', 'A24F11', 'A24F10', 'B24F11']
df = get_sql_data(cols, 'tdc')
df = df.set_index(pd.to_datetime(df['Datetime']))
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')

month_totals = df.resample('M').sum()

budget_A2400 = LP_steam_3barG['energy']/3.6*month_totals['A24F11']/month_totals['A24F06']
```

```python
##Budget 4400
rec_tags = ['44F04', 'B44F04', '44F06', 'B44F06']
df = get_sql_data(rec_tags, 'recovery')

df = df.set_index(pd.to_datetime(df['Datetime']))
df[rec_tags] = df[rec_tags].apply(pd.to_numeric, errors='coerce')






totals = df.resample('M').sum()
totals['Stripper feed ratio'] = (totals['44F04']+totals['B44F04'])/(totals['44F06']+totals['B44F06'])
totals
```

```python
## Budget recovery

cols = ['A47F01', 'B47F01', 'C47F01']
cols2 = ['A47F03', 'B47F03', 'C47F03']


effluent_composition = {
        'NMP' : 0.48,
        'Water' : 0.444, 
        'CaCl2' : 0.076
}
reflux_ratio = 0.35

df = get_sql_data(cols, 'recovery')
df = df.set_index(pd.to_datetime(df['Datetime']))
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')

df2 = get_sql_data(cols2, 'steam')
df2 = df2.set_index(pd.to_datetime(df2['Datetime']))
df2[cols2] = df2[cols2].apply(pd.to_numeric, errors='coerce')
df = df.merge(df2, left_index=True, right_index=True)




df['Total steam'] = df['A47F03']+df['B47F03']+df['C47F03']
df['Total effluent'] = df['A47F01']+df['B47F01']+df['C47F01']
df['Water feed'] = df['Total effluent']*effluent_composition['Water']

month_totals = df.resample('M').sum()


month_totals['Evap factor'] = month_totals['Water feed']/month_totals['Total steam']
```

```python
## determine recycles

cols = ['44F03','B44F03','A47F01', 'B47F01', 'C47F01']

df = get_sql_data(cols, 'recovery')
df = df.set_index(pd.to_datetime(df['Datetime']))
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')


df['Extraction feed'] = df['44F03']+df['B44F03']
df['Evap feed'] = df['A47F01']+df['B47F01']+df['C47F01']

month = df.resample('M').sum()
month
```

```python
cols = ['32K003']
production = get_sql_data(cols, 'productie')
production = production.set_index(pd.to_datetime(production['Datetime']))
production[cols] = production[cols].apply(pd.to_numeric, errors='coerce')
production = production.resample('M').sum()
month = month.merge(production['32K003'], left_index=True, right_index=True)
production
```

```python

#create new figure

fig = make_subplots(specs=[[{"secondary_y": True}]])


df = df.resample('d').sum()

#add traces

fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['Water feed']/df['Total steam'],
    mode='lines',
    name='Total 4700'))

fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['A47F01']*effluent_composition['Water']/df['A47F03'],
    mode='lines',
    name= 'A4700'))

fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['B47F01']*effluent_composition['Water']/df['B47F03'],
    mode='lines',
    name='B4700'))

fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['C47F01']*effluent_composition['Water']/df['C47F03'],
    mode='lines',
    name='C4700'))


fig.update_yaxes(title_text='Temperature (degC)')
fig.update_layout(
        template='none',
        font=dict(
                size=32,
                color="#7f7f7f"
        ),
        margin=go.layout.Margin(
            l=200,
            r=50,
            b=100,
            t=100,
            pad=4
        )
        )       

fig.update_yaxes(range=[0, 2])

py.iplot(fig)
#fig.show()
```
